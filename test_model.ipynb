```python
import numpy as np
import pickle
```

```python
model = pickle.load(open("kmeans_model.pkl", "rb"))
scaler = pickle.load(open("scaler.pkl", "rb"))
```

```python
"===== CUSTOMER SEGMENTATION SYSTEM ====="
age = float(input("Enter Age: "))
income = float(input("Enter Annual Income (k$): "))
spending = float(input("Enter Spending Score (1–100): "))


user_data = np.array([[age, income, spending]])
user_data = scaler.transform(user_data)

cluster = model.predict(user_data)

"\n This Customer Belongs To Cluster:", cluster[0]

```
